```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cs_all_emh = pd.read_excel('./cs_all_emh.xlsx', index_col =0)
corr_20012020 = pd.read_excel('./corr_20012020.xlsx', usecols ='B:E')
```

```python
cs_regre = pd.concat([cs_all_emh, corr_20012020], axis =1)
```

```python
cs_regre.rename(columns = {'year':'년도','elementary':'초등학교','middle':'중학교','high':'고등학교','all':'폐교',
                          'Elementary':'초등학생','Middle':'중학생','High':'고등학생','All':'학령인구'},inplace=True)
```

```python
cs_regre.head()
```

# 회귀분석

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
```

```python
#초등학생
model = smf.ols(formula = '폐교 ~ 초등학생', data =cs_regre)
result_ele = model.fit()
result_ele.summary()
```

- R-squared:	0.298  => 회귀모형의 설명력이 29.8%라는 뜻
- Intercept(절편)의 t값은 5.415
- P>|t|값은 0.000으로 이 회귀모형의 절편값은 유의한 것으로 나타남/ 값은 59.3569

- 초등학생의 t = -2.761, P>|t|값은 0.013, 초등학생의 계수는 -0.0089로 도출

### 회귀모형의 식은 Y = 59.3569 + (-0.0089)*초등학생 

```python
#초등학생 / 초등학교
model = smf.ols(formula = '초등학교 ~ 초등학생', data =cs_regre)
result_ele = model.fit()
result_ele.summary()
```

- R-squared:0.026  => 회귀모형의 설명력이 2.6%라는 뜻
- Intercept(절편)의 t값은 2.885
- P>|t|값은 0.010으로 이 회귀모형의 절편값은 유의한 것으로 나타남/ 값은 28.6864

- 초등학생의 t = -0.698, P>|t|값은 0.494, 초등학생의 계수는 -0.0021로 도출

### 회귀모형의 식은 Y = 28.6864 + (-0.0021)*초등학생
